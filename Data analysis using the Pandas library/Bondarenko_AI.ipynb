```python
data = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/data/adult_data.csv')
data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
       'marital-status', 'occupation', 'relationship', 'race', 'sex',
       'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
       'salary']
data.head()
```

```python
#1. Скільки чоловіків і жінок (ознака sex) представлено в цьому наборі даних?
data['sex'].value_counts()
```

```python
#2. Який середній вік (ознака age) жінок?
data[data['sex'] == ' Female']['age'].mean()
```

```python
#3. Яка частка громадян Німеччини (ознака native-country)?
(len(data[data['native-country'] == ' Germany'])/ data['native-country'].count()) * 100
```

```python
#4-5. Які середні значення і середньоквадратичні відхилення віку тих, хто отримує 
#більше 50K в рік (ознака salary) і тих, хто отримує менше 50K в рік?
data.groupby('salary')['age'].agg(['mean', 'std'])
```

```python
#6. Чи правда, що люди, які отримують більше 50k, мають як мінімум вищу освіту?
if (data[data['education'] == ' 12th']['education'].value_counts() > 0).bool() \
  | (data[data['education'] == ' 11th']['education'].value_counts() > 0).bool() \
  | (data[data['education'] == ' 10th']['education'].value_counts() > 0).bool() \
  | (data[data['education'] == ' 9th']['education'].value_counts() > 0).bool() \
  | (data[data['education'] == ' 7th-8th']['education'].value_counts() > 0).bool() \
  | (data[data['education'] == ' 5th-6th']['education'].value_counts() > 0).bool() \
  | (data[data['education'] == ' 1st-4th']['education'].value_counts() > 0).bool() \
  | (data[data['education'] == ' Some-college']['education'].value_counts() > 0).bool() \
  | (data[data['education'] == ' Some-college']['education'].value_counts() > 0).bool():
  print('Неправда. Серед людей, які отримують більше 50k, деякі не мають вищої освіти.')
else:
  print('Правда. Усі люди, які отримують більше 50k, мають як мінімум вищу освіту.')

print("\nРезультат аналізу: ", data[(data['salary'] == ' >50K')].groupby(['education'])['education'].count())
```

```python
#7. Виведіть статистику віку для кожної раси (ознака race) і кожної статі. Використовуйте groupby і describe.
#Знайдіть таким чином максимальний вік чоловіків раси Amer-Indian-Eskimo.
print('Макс. вік серед чоловіків раси Amer-Indian-Eskimo: ', data[(data['race'] == ' Amer-Indian-Eskimo') & (data['sex'] == ' Male')]['age'].max())
data.groupby(['race', 'sex'])['age'].describe()
```

```python
#8. Серед кого більше частка заробляють багато (> 50K): серед одружених або неодружених чоловіків (ознака marital-status)?
married = data[(data['sex'] == ' Male') & (data['marital-status'].isin([' Married-civ-spouse', ' Married-AF-spouse']))]
unmarried = data[(data['sex'] == ' Male') & (~data['marital-status'].isin([' Married-civ-spouse', ' Married-AF-spouse']))]

married_proportion = (married[married['salary'] == ' >50K']['salary'].count() / married['salary'].count()) * 100
unmarried_proportion = (unmarried[unmarried['salary'] == ' >50K']['salary'].count() / unmarried['salary'].count()) * 100

if married_proportion > unmarried_proportion:
 print("Частка одружених серед багатих вища, ніж серед неодружених.")
else: 
 print("Частка неодружених серед багатих вища, ніж серед одружених.")

print("\nРезультат аналізу: ")
print("Частка одружених чоловіків із заробітком >50К: ", round(married_proportion, 2), "%")
print("Частка неодружених чоловіків із заробітком >50К: ", round(unmarried_proportion, 2), "%")
```

```python
#9. Яке максимальне число годин людина працює в тиждень (ознака hours-per-week)? 
#Скільки людей працюють таку кількість годин і який серед них відсоток заробляють багато?
print("Максимальне число годин, яке людина працює в тиждень: ", data['hours-per-week'].max())

workers_max_hours = data[data['hours-per-week'] == data['hours-per-week'].max()]
count_workers_max_hours = workers_max_hours['hours-per-week'].count()
print("Кількість людей, які працюють максимальну кількість годин: ", count_workers_max_hours)

rich_workers_max_hours = workers_max_hours[workers_max_hours['salary'] == ' >50K']['salary'].count()
print("Відсоток людей серед них, які заробляють > 50K: ", round(((rich_workers_max_hours/count_workers_max_hours)*100), 2),"%")
```
