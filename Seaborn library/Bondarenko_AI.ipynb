```python
#1. Здійснити імпорт необхідних бібліотек. Імпорт Matplotlib виконати для режиму відображення графіків усередині комірок
import pandas as pd
import numpy as np
import os
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#2. Підключити ноутбук у Гугл Колаб до свого Гугл Диску
#3. Завантажити дані з файлу з інформацією про чайові (tips.csv) в Pandas DataFrame
from google.colab import drive
drive.mount('/content/gdrive')
data = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/data/tips.csv')
data.head()
```

```python
#4. Використовуючи функцію lmplot(), побудувати графік залежності розміру чайових від кількості обідаючих
sns.lmplot(x="size", y="tip", data=data)
```

```python
#5. Повторіть графіки без лінії регресії. Виділити різним кольором чи є клієнт курячим (для непарних) варіантів.
sns.lmplot(x='size', y='tip', data=data, fit_reg=False, hue='smoker')
```

```python
#6. Використовуючи box plot, показати розподіл розмірів чайових за часом замовлень – для непарних
sns.boxplot(data=data, x='time', y='tip')
```

```python
#7. Використовуючи додатковий параметр palette, на boxplot показати розподіл розмірів чайових залежно кількості обідаючих (size)
sns.boxplot(data=data, x='size', y='tip', palette='coolwarm')
```

```python
#8. Використовуючи додаткові параметри palette та hue, на boxplot
# показати розподіл розмірів чайових залежно від типу статі та факту, чи є
# клієнт курцем (окремо для кожного типу статі клієнта)

sns.boxplot(data=data, x='sex', y='tip', hue='smoker', palette='coolwarm')
```

```python
#9. Використовуючи violinplot та swarmplot, показати розподіл загального розміру рахунку залежно від часу замовлення (для непарних варіантів)
sns.violinplot(data=data, x='sex', y='tip')
plt.show()
sns.swarmplot(data=data, x='sex', y='tip')
plt.show()
sns.violinplot(data=data, x='time', y='tip')
plt.show()
sns.swarmplot(data=data, x='time', y='tip')
```

```python
#10. Використовуючи Heatmap, показати кореляцію між features датасета.

sns.heatmap(data.corr())
```

```python
#11. Використовуючи Bar plot, показати розподіл усіх категоріальних змінних датасету

sns.countplot(data=data, x='sex', hue = 'smoker')
plt.show()
sns.countplot(data=data, x='day')
```

```python
#12. Використовуючи Factor plot побудувати окремі графіки залежності розміру чайових від типу статі клієнта за різними днями тижня.
sns.factorplot(data=data, x='sex', y='total_bill', hue='day', col='day', kind='swarm')
```

```python
#13. Показати залежності між загальною сумою рахунку та розміром чайових, використовуючи pairplot

sns.pairplot(data[['total_bill', 'tip']])
```

```python
#14. Показати залежності між загальною сумою рахунку та розміром чайових з 
# урахуванням типу статі клієнта (окремим кольором), використовуючи pairplot

sns.pairplot(data[['total_bill', 'tip', 'sex']], hue='sex')
```
