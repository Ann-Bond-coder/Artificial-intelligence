```python
# 4. Створити змінну для зберігання шляху до файлу з даними (kidney_disease.csv)
# 5. Зчитати дані з файлу у датафрейм Pandas
path = os.path.join(original_path, 'kidney_disease.csv')
data = pd.read_csv(path)
```

```python
# 8. Вивести назви стовбців датафрейму
data.columns
```

```python
# 9. Створити новий набір, що містить назви стовбців, які зберігають числові дані. Для створення даного набору виконати операції двох
# зрізів на наборі назв стовбців та використати оператор додавання елементів в кінець списку для обєднання результатів зрізів в один набір.
int_float_columns = data.select_dtypes(include=['float64', 'int64']).columns
int_float_columns
```

```python
# 10. Використовуючи метод setdiff1d створити набір назв стовбців, що містять категоріальні дані, як різницю між набором з усіма назвами та набором з назвами числових стовбців.
object_columns = list(np.setdiff1d(data.columns, int_float_columns))
object_columns
```

```python
# 11. Створити екземпляр класу SimpleImputer на базі стратегії "середнє значення"
simImp = SimpleImputer(missing_values = np.nan, strategy='mean')
```

```python
# 12. Використовуючи метод fit_transform виконати заповнення тих стовбців датасету, що містять чилові дані, вивести отриманий результат
new_data = pd.DataFrame(simImp.fit_transform(data[int_float_columns]), columns=int_float_columns)
print("A new data frame:\n",new_data)
```

```python
# 13. Вивести кількість строк та стовбців (перевірити форму) отриманого на попередньому кроці результату.
new_data.shape
```

```python
# 14. Перевірити, що використання методу fit_transform не змінило оригінальний датасет. Вивести перші 5 строк оригінального датасету.
data.head()
```

```python
# 15. Використовуючи цикл по стовбцям, що містіть числові значення, оновити їх контент на відповідні значення з набору, який був отриманий в результаті трансформації. 
# Створити новий датафрейм, як копію отриманого на попередньому кроці

#16 Вивести перші 5 строк оновленого датасету.
data2 = data.copy();

for col in data2[int_float_columns].columns:
  if data2[col].dtype == 'int64' or data2[col].dtype == 'float64':
    data2[col] = new_data[col]

data2.head()
```

```python
# 17. З нового датафрейму вивести унікальні значення ознаки класс
data2['classification'].unique()
```

```python
# 18. Написати lambda-функцію для перетворення пошкодженого значення у нормальний формат ('ckd\t' це пошкоджений варіант ckd)
myF = lambda x:'ckd' if x=='ckd\t' else x
```

```python
# 19. Використовуючи метод map та реалізовану функцію, виконати перетворення значень ознаки класс
data2['classification'] = data['classification'].map(myF)
```

```python
# 20. Вивести оновлені унікальні значення ознаки класс
data2['classification'].unique()
```

```python
# 21. Використовуючи метод get_dummies виконати обробку категоріальних значень у датасеті. Вивести перші 3 рядки оновленого датасету.
data_encoded = pd.get_dummies(data2, columns=object_columns)
data_encoded.head(3)
```
